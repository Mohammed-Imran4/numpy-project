```python
s=np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],

    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],

    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
])
```

```python
# checking of each row,columns has the total 45 from 9
for i in range(0,9,3):
    for j in range(0,9,3):
        n=s[i:i+3,j:j+3]
        print(n.sum())
```

<!-- for checking of rows -->

```python
b=np.sum(s,axis=1)
for i in b:
    if i !=45:
        print("sudko is not valid")
        break
else:
    print("rows are valid")
```

# Students marks

```python
import numpy as np
# columns:[Age,maths marks,science marks]
data=np.array([
    [18,85,78],
    [19,92,88],
    [17,76,95],
    [18,65,70],
    [20,90,85]
])
```

```python
# get the shape of the matrix
data.shape
```

```python
# find the average number of the student
np.mean(data[:,0])
```

```python
# Extract Math marks of all the students
data[0:5,1:2] 
```

```python
# Find the highest science marks
np.max(data[:,2])
```

```python
# Get details of the students who scored more than 90 in maths
data[data[:,1]>90]
```

```python
# Increase math marks of all students by 5
data[:,1]+=5
data
```

```python
# find how many students are younger than 19
len(data[data[:,0]<19])
```

```python
# calculate the average marks in each subject (columns-wise mean)
np.mean(data[:,1:],axis=0)
```

```python
# Get data of student who scored at least 80 in both subject
data[(data[:,1]>=80)&(data[:,2]>=80)]
```

```python
# replace all science marks <75 with 0
data[:,2][data[:,2] < 75] = 0
data
```
